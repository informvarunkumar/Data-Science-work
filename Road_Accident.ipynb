```python
df3=df3.drop(columns=['Sl. No.'])
```

```python
df3.info()
```

```python
#remove the comma from the numeric value and convert as int format
df3["2003"]=df3["2003"].str.replace(',', '').astype("int")
```

```python
df3["2004"]=df3["2004"].str.replace(',', '').astype("int")
df3["2005"]=df3["2005"].str.replace(',', '').astype("int")
df3["2006"]=df3["2006"].str.replace(',', '').astype("int")
df3["2007"]=df3["2007"].str.replace(',', '').astype("int")
df3["2009"]=df3["2009"].str.replace(',', '').astype("int")
df3["2008"]=df3["2008"].str.replace(',', '').astype("int")
df3["2010"]=df3["2010"].str.replace(',', '').astype("int")
df3["2011"]=df3["2011"].str.replace(',', '').astype("int")
```
